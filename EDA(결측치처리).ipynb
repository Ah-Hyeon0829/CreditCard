```python
train['age'] = ((-1) * train['DAYS_BIRTH']) // 365
```

```python
train.head()
```

```python
train['age'].describe()
```

```python
bins = list(range(0,70,10))
bins_label = [str(x)+"s"  for x in bins]
```

```python
train['age'] = pd.cut(train['age'],bins, right=False, labels = bins_label[:-1])
```

```python
train.head()
```

```python
sns.countplot(x="age", data=train)
plt.title("age distribution")
plt.show()
```

```python
train['age'].value_counts()
```
